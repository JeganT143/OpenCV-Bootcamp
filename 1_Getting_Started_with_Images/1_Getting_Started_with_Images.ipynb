## Reading images using OpenCV

* It uses the <b>cv2.imread()</b> function
* We can load grayscale images, color images or load images with alpha channel 

<span style ="color:green">Function Syntax</span> <br>
<b>img_mtx = cv2.(filename[,flags])</b> <br>
<i>flags</i>
* cv2.IMREAD_GRAYSCALE or 0 => Loads image in grayscale mode
* cv2.IMREAD_COLOR or 1 => Loads a color image
* cv2.IMREAD_UNCHANGED or -1 => Loads image as such including alpha channel 

```python
# Read image as gray scale
gy_img = cv2.imread("checkerboard_18x18.png" , 0)
print(f"type of imread return value is : {type(gy_img)}")
print("\n\n")

# print the image data pixel values 
print(gy_img)
```

### Display image attributes 

```python
# size of image
print(f"size of gy_img is : {gy_img.shape}")
```

```python
# Data type of image
print(f"Data type of pixel values is : {gy_img.dtype}")
```

* matplotlib expects the image in RGB format. 
* OpenCv stores image in BGR formate
* To show correct image we need to reverse the channels 

```python
# Reverse the channels
clr_img_rev = clr_img[:,:,::-1]
plt.imshow(clr_img_rev)
```

### Splitting and Merging Color channels 

* <b>cv2.split()</b> Divides a multi-channel array into several- channel arrays 
* <b>cv2.merge()</b> Merges several arrays to make a single multi channel array
    *   All the input matrices must have the same size

```python
# Split the image into the B,G,R components
img_fr_splt = cv2.imread("New_Zealand_Lake.jpg")
b, g, r = cv2.split(img_fr_splt)

# Show the channels
plt.figure(figsize=[20,5])

plt.subplot(141); plt.imshow(r,cmap="gray"); plt.title("Red Chnnel")
plt.subplot(142); plt.imshow(g,cmap="gray"); plt.title("green Chnnel")
plt.subplot(143); plt.imshow(b,cmap="gray"); plt.title("blue Chnnel")

# Merge the individual channels into bgr
img_mrged = cv2.merge((b,g,r))

# Show the merged output 
plt.subplot(144); plt.imshow(img_mrged[:,:,::-1]); plt.title("Merged output")
```
